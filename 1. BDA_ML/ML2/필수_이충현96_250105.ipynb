```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('bank-additional.csv',sep=';')
```

```python
# y값에 의미가 정기 예금 가입하는 것에 대한 예측
display(df)
print(df.columns)
```

```python
#str 데이터로만 추출
data=df[['job','marital','education','default','housing','loan','y']]
```

```python
#불균형의 y값이다.
data['y'].value_counts()
```

```python
data['y']= data['y'].map({'yes':1,'no':0})
```

```python
## 그동안 배웠던 모델들을 성능 비교해서 평가해 보자!
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
## 범주형 데이터 전처리 원핫인코딩

categorical_columns = ['job','marital','education','default','housing','loan']
encoder = OneHotEncoder(sparse =False, drop='first')
## 인코딩 작업

encoded_features=encoder.fit_transform(data[categorical_columns])
encoded_df=pd.DataFrame(encoded_features , columns= encoder.get_feature_names_out(categorical_columns))
```

```python
## 학습 데이터 나누기
X= encoded_df
y = data['y']

X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=111, stratify=y) # 불균형 체크
```
